```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
m = 1000
X = 2 * np.random.rand(m,1)
y = 4 + 3 * X + np.random.randn(m,1)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y, marker='o', s=2.0, c='r')
plt.xlim(0,2)
plt.ylim(0,14)
plt.xlabel('X')
plt.ylabel('y', rotation=90)
plt.grid(True)
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
X_new = np.array([[0],[2]])
y_pred = lr.predict(X_new)

plt.figure(figsize=(6,4))
plt.scatter(X, y, marker='o', s=2.0, c='r')
plt.plot(X_new, y_pred, color='b')
plt.xlim(0,2)
plt.ylim(0,14)
plt.xlabel('X')
plt.ylabel('y', rotation=90)
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature

a = 2* np.random.rand(3,1)
a_ = add_dummy_feature(a)
b = 4 + 3 * a + np.random.randn(3,1)
print('a:\n', a_, '\n\n', 'b:\n', b)
```

```python
np.linalg.inv(a_.T @ a_) @ a_.T @ b
```

```python
 # Use "Normal Equation"
X_ = add_dummy_feature(X)
theta = np.linalg.inv(X_.T @ X_) @ X_.T @ y
print(theta)
```

```python
# Use "scipy least squares function"
import scipy.linalg

theta_best_svd, residuals, rank, s = scipy.linalg.lstsq(X_, y)
print(theta_best_svd)
```

```python
# Use "numpy pseudo-inverse (Moore-Penrose inverse)"

theta = np.linalg.pinv(X_) @ y
print(theta)
```
